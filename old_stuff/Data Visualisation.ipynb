```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Tune maxlimit because... we have LOTS of data
mpl.rcParams['agg.path.chunksize'] = 100000
mpl.rcParams['savefig.dpi'] = 200
mpl.rcParams['font.family'] = "Times New Roman"
mpl.rcParams['font.size'] = 30
mpl.rcParams['figure.figsize'] = 15, 10

df = pd.read_csv('consolidated.csv')
```

```python
print(df.describe(),'\n')
print(df.columns)
```

### Remove None Dates

```python
date_less = df[df['article_date']!='None']
# print(date_less.info())

date_less = date_less[(date_less['article_date']!='0201-04-10') & (date_less['article_date']!='0200-01-02')]
date_less['article_date'] = pd.to_datetime(date_less['article_date'], format='%Y-%m-%d')
date_less.head()
```

#### Plot 

Not so interesting plots of date vs length of articles.

```python
fig, ax = plt.subplots(1,1)
data_true_hyp = date_less[date_less['hyperpartisan']==1]
data_fals_hyp = date_less[date_less['hyperpartisan']==0]

ax.plot(data_fals_hyp['article_date'], data_fals_hyp['title_length'], 'o', c='g', alpha=0.5, label='False Hyperpartisan')
ax.plot(data_true_hyp['article_date'], data_true_hyp['title_length'], 'o', c='r', alpha=0.5, label='True Hyperpartisan')
ax.legend()

ax.set_title('Distribution of Title Length along the articles')
ax.set_xlabel('Date')
ax.set_ylabel('Title Length')
plt.savefig('figures/date_vs_title_len.png')
plt.show()

# date_less.plot('article_date', 'content_length', c='g', style='.')
# plt.savefig('figures/date_vs_content_len.png')
```

```python
fig, ax = plt.subplots(1,1)

ax.plot(data_true_hyp['article_date'], data_true_hyp['content_polarity']*10, 'o', c='r', alpha=0.5, label='True Hyperpartisan')
ax.plot(data_fals_hyp['article_date'], data_fals_hyp['content_polarity']*10, 'o', c='g', alpha=0.5, label='False Hyperpartisan')
ax.legend()

ax.set_title('Distribution of Sentiment along the articles')
ax.set_xlabel('Date')
ax.set_ylabel('Sentiment')
plt.savefig('figures/date_vs_sentiment.png')
plt.show()
```

```python
fig, ax = plt.subplots(1,1)

ax.plot(data_true_hyp['article_date'], data_true_hyp['content_length'], 'o', c='r', alpha=0.5, label='True Hyperpartisan')
ax.plot(data_fals_hyp['article_date'], data_fals_hyp['content_length'], 'o', c='g', alpha=0.5, label='False Hyperpartisan')
ax.legend()

ax.set_title('Distribution of Content Length along the articles')
ax.set_xlabel('Date')
ax.set_ylabel('Content Length')
plt.savefig('figures/date_vs_content_len.png')
plt.show()
```

Some more plots, depicting ground truth.

```python
hyper_part = date_less[date_less['hyperpartisan']==1]
normal = date_less[date_less['hyperpartisan']==0]
ax = hyper_part.plot('article_date', 'orientation', c='r', style='.')
normal.plot('article_date', 'orientation', c='g', ax=ax, style='.')
ax.legend(['true hyperpartisan', 'false hyperpartisan'])
# plt.savefig('figures/date_vs_orientation.png')
```

```python
hyper_part['content_polarity'] = 10*hyper_part['content_polarity']
normal['content_polarity'] = 10*normal['content_polarity']
ax = hyper_part.plot('article_date', 'content_polarity', c='r', style='.', alpha=0.1)
normal.plot('article_date', 'content_polarity', c='g', style='.', ax=ax, alpha=0.1)
ax.legend(['true hyperpartisan', 'false hyperpartisan'])
# plt.savefig('figures/date_vs_sentiment.png')
```
